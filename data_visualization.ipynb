```python
import load_images
import matplotlib.pyplot as plt
```

```python
X_train, Y_train, X_validation, Y_validation, X_test, Y_test = load_images.getDataset()
```

```python
classes = ['angry', 'disgusted', 'fearful', 'happy', 'neutral', 'sad', 'surprised']
# classes_to_num = {'angry': 0, 'disgusted': 1, 'fearful': 2, 'happy': 3, 'neutral': 4, 'sad': 5, 'surprised': 6}
count = load_images.getDatasetCount()
print(count)
plt.bar(classes,count)
```

```python
import numpy as np
unique1, counts1 = np.unique(Y_train, return_counts=True)
print(np.asarray((unique1, counts1)))
```

```python
unique2, counts2 = np.unique(Y_validation, return_counts=True)
print(np.asarray((unique2, counts2)))
```

```python
print(np.asarray((unique2, counts2/counts1)))
```

```python
d = {}
for i in Y_train:
    if i not in d:
        d[i] = 0
    else:
        d[i] +=1
print(d)
```

```python
# Data Visualization grayscale
fig, axes = plt.subplots(3, 3, figsize=(10, 10)) 
for i in range(9):
    img = X_train[i]
    axes[i//3, i%3].imshow(img) # show the image
    label = Y_train[i] 
    axes[i//3, i%3].set_title(f'Label: {label}') 
```
